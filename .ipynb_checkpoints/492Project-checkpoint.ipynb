```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import xgboost
import numpy as np
import graphviz
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import classification_report
import seaborn as sns
```

```python
heart = pd.read_csv("heart.csv")
X = heart.drop("output", axis=1)
y = heart["output"]
```

```python
scores = [0]
for i in range(1,14):
    X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
    kbest = SelectKBest(k=i)
    kbest.fit(X_train, y_train)
    features = kbest.get_feature_names_out()
    X_train = X_train[features]
    X_test = X_test[features]
    xgb =  xgboost.XGBClassifier(learning_rate=0.1, max_depth=1, min_child_weight=9, n_estimators=100, n_jobs=1, subsample=0.7500000000000001, verbosity=0)
    xgb.fit(X_train, y_train)
    curr_score = xgb.score(X_test,y_test) 
    scores.append(curr_score)
```

```python
best_score = max(scores)
best_k = scores.index(best_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
kbest = SelectKBest(k=best_k)
kbest.fit(X_train, y_train)
features = kbest.get_feature_names_out()
X_train = X_train[features]
X_test = X_test[features]
```

```python
xgb =  xgboost.XGBClassifier(learning_rate=0.1, max_depth=1, min_child_weight=9, n_estimators=100, n_jobs=1, subsample=0.75, verbosity=0)
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```

```python
for i in range(5):
    xgboost.plot_tree(xgb,num_trees=i)
```

```python
y_pred = xgb.predict(X_test)
proba = xgb.predict_proba(X_test)
confidence = []

for i in range(len(y_pred)):
    if y_pred[i] == 0:
        confidence.append(proba[i,0])
    else:
        confidence.append(proba[i,1])
predictions = pd.DataFrame({"Prediction":y_pred, "Confidence":confidence})
predictions.to_csv("predictions.csv")
print(predictions.head())
```

```python
report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
report.drop("support", inplace=True)
sns.heatmap(report, annot=True, cmap="rocket_r", cbar=False)
plt.title("Classification Report")
plt.savefig("visualizations/classificationreport.png")
```
