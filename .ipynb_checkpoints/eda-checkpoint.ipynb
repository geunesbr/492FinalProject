```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

```python
heart = pd.read_csv("heart.csv")
```

```python
heart
```

```python
sns.pairplot(heart)
plt.title("Heart Data Pairplot")
plt.savefig("visualizations/heartpairplot.png")
```

```python
plt.scatter(heart['age'],heart['thalachh'], c=heart["output"])
plt.title("Heart Rate by Age Colored for Heart Attack Risk")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.savefig("visualizations/heartratebyage.png")
plt.show()
```

```python
plt.hist(heart['age'])
plt.title("Heart Attack Frequency by Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("visualizations/agehist.png")
plt.show()
```

```python
def transform(X, n): #scale and perform pca
    scaler = StandardScaler()
    scaler.fit(X)
    X_transform = scaler.transform(X)
    pca = PCA(n_components=n)
    pca.fit(X_transform)
    X_pca = pca.transform(X_transform)
    return X_pca
```

```python
X = heart.drop("output", axis=1)
y = heart["output"]
```

```python
X_pca = transform(X, 3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=y)
ax.set_xlabel("PC1", fontsize=10)
ax.set_ylabel("PC2", fontsize=10)
ax.set_zlabel("PC3", fontsize=10, labelpad=.01)
ax.set_title("Principal Components Shaded for Heart Attack Risk");
plt.savefig("visualizations/pca.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=y)
ax.view_init(270,270)
ax.set_xlabel("PC1", fontsize=10)
ax.set_ylabel("PC2", fontsize=10)
ax.set_zlabel("PC3", fontsize=10)
ax.set_title("Principal Components Shaded for Heart Attack Risk (Alternate Angle)");
plt.savefig("visualizations/pca2.png")
plt.show()
```

```python
for k in ['linear', 'poly', 'rbf', 'sigmoid']: # fit svm with each of four kernels
    clf = svm.SVC(kernel=k)
    clf.fit(X_pca, y)
    print(k)
    print(clf.score(X_pca,y))
```

```python
from tpot import TPOTClassifier
pipeline_optimizer = TPOTClassifier(generations=10)
pipeline_optimizer.fit(X, y)
print("Baseline:",pipeline_optimizer.score(X, y))
```

```python
pipeline_optimizer.export('tpot_exported_pipeline.py')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
kbest = SelectKBest(k=8)
kbest.fit(X_train, y_train)
features = kbest.get_feature_names_out()
X_train = X_train[features]
X_test = X_test[features]
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
et = ExtraTreesClassifier()
et.fit(X_train, y_train)
et.score(X_test, y_test)
```
